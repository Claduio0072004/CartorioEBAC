```python
# Criar um dicionário com a tradução das colunas
traducao_colunas = {
    'work_year': 'ano',
    'experience_level': 'nivel_experiencia',
    'employment_type': 'tipo_emprego',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda_salario',
    'salary_in_usd': 'salario_em_usd',
    'employee_residence': 'residencia_funcionario',
    'remote_ratio': 'percentual_remoto',
    'company_location': 'localizacao_empresa',
    'company_size': 'tamanho_empresa'
}

# Renomear as colunas do DataFrame
df.rename(columns=traducao_colunas, inplace=True)

# Exibir as colunas renomeadas para verificar
print(df.columns)
```

```python
# Criar um dicionário para traduzir as categorias de nível de experiência
nivel_experiencia = {
    'SE': 'Sênior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
}

# Traduzir as categorias na coluna 'nivel_experiencia'
df['nivel_experiencia'] = df['nivel_experiencia'].replace(nivel_experiencia)

# Exibir a contagem dos valores traduzidos para verificar
df['nivel_experiencia'].value_counts()
```

```python
# Criar um dicionário para traduzir as categorias de tipo de emprego
tipo_emprego = {
    'FT': 'Integral',
    'CT': 'Contrato',
    'PT': 'Parcial',
    'FL': 'Freelancer'
}

# Traduzir as categorias na coluna 'tipo_emprego'
df['tipo_emprego'] = df['tipo_emprego'].replace(tipo_emprego)

# Exibir a contagem dos valores traduzidos para verificar
df['tipo_emprego'].value_counts()
```

```python
# Criar um dicionário para traduzir as categorias de tamanho de empresa
tamanho_empresa = {
    'M': 'Médio',
    'L': 'Grande',
    'S': 'Pequeno'
}

# Traduzir as categorias na coluna 'tamanho_empresa'
df['tamanho_empresa'] = df['tamanho_empresa'].replace(tamanho_empresa)

# Exibir a contagem dos valores traduzidos para verificar
df['tamanho_empresa'].value_counts()
```

```python
# Criar um dicionário para traduzir as categorias de percentual remoto
remoto = {
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
}

# Traduzir as categorias na coluna 'percentual_remoto'
df['percentual_remoto'] = df['percentual_remoto'].map(remoto)

# Exibir a contagem dos valores traduzidos para verificar
df['percentual_remoto'].value_counts()
```

```python
troca_remoto = {
    'percentual_remoto': 'remoto'
}

df.rename(columns=troca_remoto, inplace=True)

print(df.columns)
```

```python
import numpy as np

dfs = pd.DataFrame({'Nome': ["Claudio", "Marcelo", "Ana", "Lana", "Maria", "Lionel"], 'Salario': [np.nan, 2300, np.nan, 4000, 1950, np.nan]})

dfs['salario_meida'] = dfs['Salario'].fillna(dfs['Salario'].mean().round(2))

dfs['salario_meidana'] = dfs['Salario'].fillna(dfs['Salario'].median())

dfs
```

```python
dft = pd.DataFrame({'Dia': ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta'], 'Temperatura': [30, np.nan, np.nan, 28, 27]})

dft['preenchido_ffill'] = dft['Temperatura'].ffill()

dft
```

```python
dft = pd.DataFrame({'Dia': ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta'], 'Temperatura': [30, np.nan, np.nan, 28, 27]})

dft['preenchido_bfill'] = dft['Temperatura'].bfill()

dft
```

```python
dfc = pd.DataFrame({'Nome': ["Claudio", "Marcelo", "Ana", "Lana", "Maria", "Lionel"], 'Cidade': ["São Paulo", np.nan, "Curitiba", "Belém", np.nan, "Rio de Janeiro"]})

dfc['cidade_preenchida'] = dfc['Cidade'].fillna('Não Informado')

dfc
```

```python
import pycountry

def iso2_to_iso3(code):
    try:
        return pycountry.countries.get(alpha_2=code).alpha_3
    except:
        return None

dfl['residencia_iso3'] = dfl['residencia_funcionario'].apply(iso2_to_iso3)

df_ds = dfl[dfl['cargo'] == 'Data Scientist']
media_ds_pais = df_ds.groupby('residencia_iso3')['salario_em_usd'].mean().reset_index()

fig = px.choropleth(media_ds_pais,
                    locations='residencia_iso3',
                    color='salario_em_usd',
                    color_continuous_scale='rdylgn',
                    title='Salário médio de Cientista de Dados por país',
                    labels={'salario_em_usd': 'Salário médio (USD)', 'residencia_iso3': 'País'})

fig.show()
```
